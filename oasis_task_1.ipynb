```python
# View the first few rows of both dataframes
print("Menu DataFrame:")
print(menu_df.head())

print("\nSales DataFrame:")
print(sales_df.head())
```

```python
# 📋 First clean column names (remove spaces)
sales_df.columns = sales_df.columns.str.strip()
menu_df.columns = menu_df.columns.str.strip()

# 🏷 Set the correct matching columns
sales_column = 'Product Category'   # from sales_df
menu_column = 'Item'                # from menu_df

# 🛡️ Safety check
if sales_column not in sales_df.columns:
    raise ValueError(f"'{sales_column}' not found in sales_df columns: {sales_df.columns.tolist()}")

if menu_column not in menu_df.columns:
    raise ValueError(f"'{menu_column}' not found in menu_df columns: {menu_df.columns.tolist()}")

# 🔗 Merge datasets
merged_df = pd.merge(sales_df, menu_df, left_on=sales_column, right_on=menu_column, how='left')

# ✅ Preview
print("\n✅ Merged Dataset Preview:")
print(merged_df.head())
```

```python
# See data info
print(merged_df.info())

# Check missing values
print(merged_df.isnull().sum())

# See basic statistics
print(merged_df.describe())
```

```python
merged_df['Date'] = pd.to_datetime(merged_df['Date'])  # Make sure Date is datetime
merged_df.set_index('Date', inplace=True)

monthly_sales = merged_df['Total Amount'].resample('M').sum()

plt.figure(figsize=(12,6))
plt.plot(monthly_sales, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid()
plt.show()
```

```python
top_products = merged_df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_products.plot(kind='bar')
plt.title('Top 10 Selling Products')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Select only numeric columns
numeric_df = merged_df.select_dtypes(include=['float64', 'int64'])

# Now plot heatmap
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Numeric Features Only)')
plt.show()
```
